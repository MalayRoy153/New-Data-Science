```python
import pandas as pd
```

```python
df = pd.read_csv('house price.csv')
```

DATA PREPROCESSING

```python
df
```

```python
df.info()
```

```python
df = df.drop(columns=['street', 'city',
       'statezip', 'country','date'])
```

```python
df.columns
```

```python
df = df[(df['sqft_basement'] != 0) & (df['yr_renovated'] != 0)]
```

```python
df
```

```python
df.info()
```

DATA SPLITING

```python
from sklearn.model_selection import train_test_split

X = df.drop(['price'], axis=1)
y = df['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
train_df = X_train.join(y_train)
```

```python
train_df
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Chceking relation between the datas
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_df.corr(numeric_only=True), annot=True, cmap="YlGnBu")
```

LINEAR REGRESSOR

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_s= scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train_s,y_train)
```

```python
y_pred = lr.predict(X_test_s)
```

```python
#To get R**2 score between 0 and 1 , closer to 1 , the better the value
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y_test, y_pred)
r2
```

```python
#To get RMSE AND MSE AND RMAE AND MAE   /// Lower the values better the prediction model
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
from sklearn.metrics import mean_squared_error
```

```python
# Mean Squared Error
```

```python
MSE = mean_squared_error(y_test, y_pred)
MSE
```

```python
# Root Mean Squared Error
```

```python
RMSE = np.sqrt(MSE)
RMSE
```

```python
#Mean Absolute Error
```

```python
MAE = mean_absolute_error(y_test, y_pred)
MAE
```

```python
#Root Mean Absolute Error
```

```python
RMAE = np.sqrt(MAE)
RMAE
```

RANDOM FOREST REGRESSOR

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators=300, criterion='friedman_mse', random_state=42)
rf.fit(X_train_s, y_train)
```

```python
rfy_pred = rf.predict(X_test_s)
```

```python
#To get R**2 score between 0 and 1 , closer to 1 , the better the value
r2 = r2_score(y_test, rfy_pred)
r2
```

```python
# Mean Squared Error
MSEs = mean_squared_error(y_test, rfy_pred)
MSEs
```

```python
# Root Mean Squared Error
RMSEs = np.sqrt(MSEs)
RMSEs
```

```python
#Mean Absolute Error
MAEs = mean_absolute_error(y_test, rfy_pred)
MAEs
```

```python
# Root Mean Absolute Error
RMAEs = np.sqrt(MAEs)
RMAEs
```
